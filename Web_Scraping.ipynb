```python
url ="https://www.sc.fsu.edu/people/administration"
```

```python
##sending request

response=requests.get(url)
```

```python
response
```

```python
response.raise_for_status()
```

```python
##parse html content
domobjct=BeautifulSoup(response.text, 'html.parser')
```

```python
domobjct
```

```python


## we have to extract the data from dom object so creating / declaring one data set
data=[]
```

```python
## serach the table data in dom object and extract it
tableData=domobjct.find("table")
```

```python
## we got entire <table> tag data from dom obejct
## need to extract the exact data from table object

if tableData:
  rows=tableData.findAll("tr");
  for row in rows:
    cols=row.findAll("td");
    cols=[col.text.strip() for col in cols]
    data.append(cols)
```

```python
data
```

```python
cleaned_data = [[item.replace("This email address is being protected from spambots. You need JavaScript enabled to view it.", "") for item in sublist] for sublist in data]
cleaned_data = [[item.replace("\n", "") for item in sublist] for sublist in cleaned_data]

cleaned_data
```
